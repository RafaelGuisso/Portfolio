```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
import math
from matplotlib.pyplot import figure

def odeEuler(f,y0,t):
  '''
    Parametros

    f : funcao
        lado direito da equacao diferencial y'=f(t,y), y(t_0)=y_0
    y0 : Valor inicial
    t : (x inicial, x final, intervalo entre os pontos)
  '''

  y = np.zeros(len(t))
  y[0] = y0
  for n in range(0,len(t)-1):
      y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
  return y
```

```python
h = 0.1
t = np.arange(0, 1 + h, h)
y0 = -1
f = lambda y,t: np.exp(-t)
y = odeEuler(f,y0,t)
t_real = np.linspace(0,1,100)
y_real = -np.exp(-t_real)

figure(figsize=(11, 11), dpi=80)
plt.plot(t,y,'b.-',t_real,y_real, 'r-')
plt.legend(['Euler','Real'])
plt.grid(True)
plt.axis([0,1,-1,0])
plt.show()
print (y)
```

```python
h = 0.1
t = np.arange(2, 3 + h, h)
y0 = np.e
f = lambda y,t: (y*np.log(y))/t
y = odeEuler(f,y0,t)
t_real = np.linspace(2,3,100)
y_real = np.exp(t_real/2)

figure(figsize=(11, 11), dpi=80)
plt.plot(t,y,'b.-',t_real,y_real, 'r-')
plt.legend(['Euler','Real'])
plt.grid(True)
plt.axis([2,3,2,5])
plt.show()
print (y)
```
