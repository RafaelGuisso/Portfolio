```python
# Cria função com 'x' e 'y' independentes em função de 'z'

def func(X, b0, b1, b2, b3):
  X = hi, hi1, BEi, T, R
  Ldi = (R*(hi1-hi))**(1/2)
  fator1 = (BEi/hi1)
  fator2 = (BEi/Ldi)
  fator3 = (hi1/R)
  CH = np.exp(b0*(fator1**b1)*(fator2**(b2*fator1))*(fator3**(b3*fator1)))
  gT = 1 + 0.002*(T - 1250)
  return (((hi1/hi)**CH)-1)*BEi*gT
    
# gT = 1 + 0,002*(T - 1250)

# Define valores para 'x' e 'y' e simula curva 'actual' para 'z' (simulação real, com ruídos)
hi = db['hi-1_6p']
hi1 = db['hi-1_5p']
BEi = db['Bei_H_1p']
T = db['T_6p']
R = db['R_top']
DeltaB =  db ['DeltaBhi_actual_6p']

#b0, b1, b2, b3 = -1.7, 0.38, 0.01, -0.01
b0, b1, b2, b3 = -1.7, 0.38, 0.01, -0.01

# Parametriza equação a partir de dados iniciais dos coeficientes e mostra a covariância
#p0 =   -1.7, 0.38, 0.01, -0.01
p0 =   -1.7, 0.38, 0.01, -0.01
popt, pcov = curve_fit( func, (hi, hi1, BEi, T, R), DeltaB, p0, bounds=([-1000, -100, -100, -100], [100, 100, 100, 100]))
popt, pcov
```
